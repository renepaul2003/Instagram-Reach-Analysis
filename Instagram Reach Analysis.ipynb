###  Training and Testing the Dataset

```python
#Splitting the data 
from sklearn.model_selection import train_test_split
#assgining the datas into x and y 
x=data.drop(["Follows"],axis=1)
y=data["Follows"]
#shapes of the x and y
x.shape,y.shape
```

```python
#training and testing the data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
#identifying the shape of the xtrain ,xtest,ytrain,ytest
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

### Model Devolpment

There are so many state-of-the-art ML models available in academia but some model fits better to some problem while some fit better than other. So, to make this decision we split our data into training and validation data. Then we use the validation data to choose the model with the highest performance

In this regression problem ,we use these models

* LinearRegression
* SupportVectorRegressor(SVR)
* RandomForestRegressor
* DecisionTreeRegressor
* GradientBoostingRegressor
* KNeighborsRegressor


To predict the accuracy we will use the accuracy score function from scikit-learn library.

###  1. LinearRegression

```python
# importing the model LinearRegression
from sklearn.linear_model import LinearRegression
#importing r2_score and cross_val_score for predicting the score
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
#fitting the xtrain and ytrain into the model LineaRegression
lr.fit(xtrain,ytrain)
```

```python
# predicting the training and testing the dataset
yprdtrain=lr.predict(xtrain)
yprdtest=lr.predict(xtest)
# predicting the score of the model LinearRegression
print(r2_score(ytrain,yprdtrain))  # Traning Accuracy
print(r2_score(ytest,yprdtest))    #testing Acuracy
print(cross_val_score(lr,x,y,cv=5,).mean()) 
```

### 2. SVR(Support Vector Regressor)

```python
# importing the model SVR
from sklearn.svm import SVR
sr=SVR()
# fitting xtrain and ytrain into model SVR
sr.fit(xtrain,ytrain)
```

```python
yprdtrain1=sr.predict(xtrain)
yprdtest1=sr.predict(xtest)
# predicting the score of the model
print(r2_score(ytrain,yprdtrain1))
print(r2_score(ytest,yprdtest1))
print(cross_val_score(sr,x,y,cv=5,).mean())
```

###  3. Random Forest Regressor

```python
# importing the model RandomForest Regressor
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
# fitting xtrain and ytrain into the model randomforestregressor
rf.fit(xtrain,ytrain)
```

```python
yprdtrain2=rf.predict(xtrain)
yprdtest2=rf.predict(xtest)
# predicting the score of the model 
print(r2_score(ytrain,yprdtrain2))
print(r2_score(ytest,yprdtest2))
print(cross_val_score(rf,x,y,cv=5,).mean())
```

### 4.  DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
# fitting  xtrain and ytrain into model
dt.fit(xtrain,ytrain)
```

```python
yprdtrain3=dt.predict(xtrain)
yprdtest3=dt.predict(xtest)
# predicting the score of the model
print(r2_score(ytrain,yprdtrain3))
print(r2_score(ytest,yprdtest3))
print(cross_val_score(dt,x,y,cv=5,).mean())
```

###   5. GradientBoostingRegressor

```python
# importing the model GradientBoostingRegressor 
from sklearn.ensemble import GradientBoostingRegressor
gb=GradientBoostingRegressor()
#fitting the xtrain and ytrain into models
gb.fit(xtrain,ytrain)
```

```python
yprdtrain4=gb.predict(xtrain)
yprdtest4=gb.predict(xtest)
# predicting the score of the model
print(r2_score(ytrain,yprdtrain4))
print(r2_score(ytest,yprdtest4))
print(cross_val_score(gb,x,y,cv=5,).mean())
```

### 6. KNeighborsRegressor

```python
# importing the model KNeighborsRegressor
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
# fitting xtrain and ytrain into model
kn.fit(xtrain,ytrain)
```

```python
yprdtrain5=kn.predict(xtrain)
yprdtest5=kn.predict(xtest)
# predicting the score of the model
print(r2_score(ytrain,yprdtrain5))
print(r2_score(ytest,yprdtest5))
print(cross_val_score(kn,x,y,cv=5,).mean())
```
